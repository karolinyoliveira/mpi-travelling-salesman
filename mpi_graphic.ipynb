```python
# Dependências
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Dados

# Quantia de execuções tomadas para a média de tempo
num_attempts = 32

# Tamanhos do grafo (em nós)
sizes = np.arange(2, 20)
print("sizes = {}".format(sizes))

# Tempos de execução (em segundos)
times = np.array([i**3 for i in range(2,20)])
print("times = {}".format(times))
```

```python
# Gráfico
plt.style.use("bmh")
plt.figure(figsize=(6,6))
plt.plot(sizes, times)
plt.xticks(sizes)
plt.ylabel("tempo médio de resposta (segundos)")
plt.xlabel("tamanho do grafo (número de nós)")
plt.show()
```
